```python
# Import packages
import numpy as np
import pandas as pd
```

```python
# Import CVS files into dataframes

dim_customers = pd.read_csv("dim_customers.csv")
dim_date = pd.read_csv("dim_date.csv")
dim_products = pd.read_csv("dim_date.csv")
dim_targets_orders = pd.read_csv("dim_targets_orders.csv")
fact_order_lines = pd.read_csv("fact_order_lines.csv")
fact_orders_aggregate = pd.read_csv("fact_orders_aggregate.csv")
```

```python
# Check for data shapes
print("dim_customers table---\n")
print(dim_customers.shape)

print("\ndim_date table---\n")
print(dim_date.shape)

print("\ndim_products table---\n")
print(dim_products.shape)

print("\ndim_targets_orders table---\n")
print(dim_targets_orders.shape)

print("\nfact_order_lines table---\n")
print(fact_order_lines.shape)

print("\nfact_orders_aggregate---\n")
print(fact_orders_aggregate.shape)
```

```python
# Check for top 5 data
print("dim_customers table Data---\n")
print(dim_customers.head())

print("\ndim_date table Data---\n")
print(dim_date.head())

print("\ndim_products table Data---\n")
print(dim_products.head())

print("\ndim_targets_orders table Data---\n")
print(dim_targets_orders.head())

print("\nfact_order_lines table Data---\n")
print(fact_order_lines.head())

print("\nfact_orders_aggregate Data---\n")
print(fact_orders_aggregate.head())
```

```python
# Check for missing values
print("dim_customers table missing values---\n")
print(dim_customers.isna().sum())

print("\ndim_date table missing values---\n")
print(dim_date.isna().sum())

print("\ndim_products table missing values---\n")
print(dim_products.isna().sum())

print("\ndim_targets_orders table missing values---\n")
print(dim_targets_orders.isna().sum())

print("\nfact_order_lines table missing values---\n")
print(fact_order_lines.isna().sum())

print("\nfact_orders_aggregate missing values---\n")
print(fact_orders_aggregate.isna().sum())
```
